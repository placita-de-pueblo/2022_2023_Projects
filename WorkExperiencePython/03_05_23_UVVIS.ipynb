```python
import numpy as np 
import os
import matplotlib.pyplot as plt
os.chdir("C:/Users/claud/OneDrive/Escritorio/Praktikum_Python/UVVIS")
```

```python
datenBaseRB = np.loadtxt("BaseRoteBeete_Absorbance_13-54-42-824.txt")
datenNeutralRB = np.loadtxt("NeutralRoteBeete_Absorbance_13-52-01-661.txt")
datenSaureRB = np.loadtxt("SaureRoteBeete_Absorbance_13-57-03-823.txt")
datenEthanolOrange = np.loadtxt("EthanolOrange_Absorbance_14-03-03-555.txt")
datenChl = np.loadtxt("Chlamydomonas_Absorbance_14-34-58-487.txt")
datenSyn = np.loadtxt("Synechosistus_Absorbance_14-32-29-483.txt")
```

```python
np.shape(datenBaseRB)
np.shape(datenNeutralRB)
np.shape(datenSaureRB)
xBase = datenBaseRB[:,0]
yBase = datenBaseRB[:,1]
yNeutral = datenNeutralRB[:,1]
ySaure = datenSaureRB[:,1]
n=3
fs=30
plt.figure(figsize=(40,30))
plt.plot(xBase/100, yBase, c="midnightblue", lw=n+1, ls="solid", label="Base")
plt.plot(xBase/100, yNeutral, c="forestgreen", lw=n+1, ls="solid", label="Neutral")
plt.plot(xBase/100, ySaure, c="firebrick", lw=n+1, ls="solid", label="Sure")
plt.xlabel("Wellenlnge in nm", fontsize = fs+5)
plt.ylabel("Intensitt", fontsize = fs+5)
plt.legend(loc=2, fontsize=fs+10)
plt.xticks(fontsize = fs-5)
plt.yticks(fontsize = fs-5)
plt.xlim(3, 7)
plt.ylim(0,2)
```

```python
np.shape(datenEthanolOrange)
x = datenEthanolOrange[:,0]
yEO = datenEthanolOrange[:,1]
n = 3
fs = 30
plt.figure(figsize=(40,30))
plt.plot(x/100, yEO, c="coral", lw=n+1, ls="solid", label="Ethanol")
plt.xlabel("Wellenlnge in nm", fontsize = fs+5)
plt.ylabel("Intensitt", fontsize = fs+5)
plt.legend(loc=2, fontsize=fs+10)
plt.xticks(fontsize = fs-5)
plt.yticks(fontsize = fs-5)
plt.xlim(3, 7)
plt.ylim(0,7)
```

```python
np.shape(datenChl)
np.shape(datenSyn)
x = datenChl[:,0]
yChl = datenChl[:,1]
ySyn = datenSyn[:,1]
n = 3
fs = 30
plt.figure(figsize=(40,30))
plt.plot(x/100, yChl, c="seagreen", lw=n+1, ls="solid", label="Chlamydomonas")
plt.plot(x/100, ySyn, c="limegreen", lw=n+1, ls="solid", label="Synechosistus")
plt.xlabel("Wellenlnge in nm", fontsize = fs+5)
plt.ylabel("Intensitt", fontsize = fs+5)
plt.legend(loc=2, fontsize=fs+10)
plt.xticks(fontsize = fs-5)
plt.yticks(fontsize = fs-5)
plt.xlim(3, 7)
plt.ylim(-0.2,0.8)
```
