### Importieren der Dateien, sowie der Funktionen, die im Folgenden benutzt werden

```python
import numpy as np 
import os
import matplotlib.pyplot as plt
os.chdir("C:/Users/claud/OneDrive/Escritorio/Praktikum_Python/IR")

datenPusten = np.loadtxt("Pusten.txt")
datenZelle = np.loadtxt("Zelle.txt")
datenGrueneFlasche = np.loadtxt("GrueneFlasche.txt")
datenFingerabdruck = np.loadtxt("Fingerabdruck.txt")
datenFeineGraueSchicht = np.loadtxt("FeineGraueSchicht.txt")
```

## Methan

```python
np.shape(datenZelle)
x = datenZelle[:,0]
yZelle = datenZelle[:,2]
Zelle = "darkorange"
plt.figure(figsize=(5,2.5))
plt.plot(x/100, yZelle, c="cornflowerblue", lw=1, ls="solid", label="Methan")
plt.xlabel("Wellenzahl in 1/cm", fontsize = 10)
plt.ylabel("Intensit√§t", fontsize = 10)
plt.legend(loc=2, fontsize=10)
plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)
plt.xlim(3200, 2800)
plt.ylim(0.15,0.55)
plt.axvline(3016, 0, 0.8, lw=1, color = Zelle)
plt.text(3017, 0.49, "CH\nstretch", horizontalalignment = "center", fontsize= 8)
plt.text(3100, 0.3, "Rotations-\nschwingungen", horizontalalignment = "center", fontsize= 8)
plt.text(2950, 0.3, "Rotations-\nschwingungen", horizontalalignment = "center", fontsize= 8)
plt.savefig("Methan.png", bbox_inches="tight")
```

## Ethanol

```python
np.shape(datenGrueneFlasche)
Ethanol = "salmon"
n = 30
yGF = datenGrueneFlasche[:,2]
plt.figure(figsize=(40,30))
plt.plot(x/100, yGF, c="mediumpurple", lw=3, ls="solid", label="Ethanol")
plt.xlabel("Wellenzahl in 1/cm", fontsize = n+5)
plt.ylabel("Intensit√§t", fontsize = n+5)
plt.legend(loc=2, fontsize=n+10)
plt.xticks(fontsize=n-5)
plt.yticks(fontsize=n-5)
plt.xlim(4000,820)
plt.ylim(-0.01,0.15)
plt.axvline(2971, 0, 0.72, lw=3, color=Ethanol)
plt.text(2975, 0.11, "CH\nstretch", horizontalalignment = "center", fontsize = n)
plt.axvline(1393, 0, 0.30, lw=3, color=Ethanol)
plt.text(1090, 0.135, "Backbone\nvibration", horizontalalignment = "center", fontsize= n)
plt.axvline(1066, 0, 0.8, lw = 3,color=Ethanol)
plt.text(1380, 0.055, "CH$_{3}$\nstretch", horizontalalignment = "center", fontsize = n)
plt.savefig("Ethanol.png")
```

## Atemluft
